```python
def related_str_table(class_le,name):
    related_date = {}
    for cl in class_le.classes_:
        related_date.update({cl:int(class_le.transform([cl])[0])})
    with open("data/" + name + ".txt","w") as f:
        json.dump(related_date,f,indent=4)
```
