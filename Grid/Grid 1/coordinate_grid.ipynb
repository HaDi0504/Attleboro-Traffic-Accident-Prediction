# python 坐标的网格化处理
    输入：坐标经纬度，网格的行数和列数  
    输出：坐标的网格区域ID  

```python
import pandas as pd
import random
```

```python
# Area
LON1 = 41.893095 
LON2 = 41.986257
LAT1 = -71.381891
LAT2 = -71.269201
```

```python
data=pd.read_csv('final.csv')
```

```python
def generalID(lon,lat,column_num,row_num):
    if lon <= LON1 or lon >= LON2 or lat <= LAT1 or lat >= LAT2:
        return -1
    column = (LON2 - LON1)/column_num
    row = (LAT2 - LAT1)/row_num
    return int((lon-LON1)/column)+ 1 + int((lat-LAT1)/row) * column_num
```

```python
#x times x
data['label'] = data.apply(lambda x: generalID(x['lon'], x['lat'],25,25), axis = 1)
```
